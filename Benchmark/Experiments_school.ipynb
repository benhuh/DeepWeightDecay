# Different choices of $m$

```python
def get_t(m, d, r, factor):
	return math.ceil(r * (d - r) / (m - r) * factor)
```

```python
get_t(8, 27, 2, 1.25)
```

## Explaining the panels

In the figures below, the training loss on the left is caludated using the same dataset used to calculate $C_{opt}$. The test and train mse on the right calculate $C_{opt}$ with 4 data points which will be excluded from the set calculating mse. When $m = 4$ training mse is not calculated on the right most panel.

## m = 4

```python
prob = RealDataProb()
prob.generate_data(m_train = 4, t_train = 23, m_val = 25, t_val = 23, seed = 5)
print(
    "Training config: m: %d, t: %d \n"%(prob.X.shape[1], prob.X.shape[0]),
    "Testing config: m: %d, t: %d"%(prob.X_val.shape[1], prob.X_val.shape[0])
)
output, history = apply_alt_min_gd_wd(prob, 2000, (0.01, 0.01), num_layer = 2, wide = True, if_output_hist = True, weight_decay = 0.001, epoch_div = 3)
_ = plot_all(history, 'losses', 'sing_val', 'test_mse', 'train_mse')
```
